```python
df = pd.read_csv('japan_df_province')
```

```python
df['Date'] = pd.to_datetime(df['Captured Time'], format='%Y-%m-%d %H:%M:%S')#.dt.year
df.drop('Captured Time', axis=1, inplace=True)
```

```python
df = df[df['Date'].notnull()]
df = df.sort_values('Date')
```

```python
df = df.groupby([pd.Grouper(key='Date', freq='D'), 'Province']).mean().reset_index()
```

```python
df['Date'] = df['Date'].dt.year
```

```python
df = df.groupby(['Date', 'Province']).mean().reset_index()
```

```python
df = df[['Date','Province','Value']]
```

```python
dates = df['Date'].unique()
provinces = df['Province'].unique()
```

```python
dates_t, province_t = [], []
for i in dates:
    for j in provinces:
        dates_t.append(i)
        province_t.append(j)
        
```

```python
def get_diff(v1, v2):
    return abs(v1 - v2)
```

```python
over_years_matrix = {}
```

```python
over_years_matrix = {}
for year in dates:
    matrix = [[0 for x in range((len(provinces)))] for y in range((len(provinces)))] 
    time_df = df[df['Date'] == year]
    for province_1 in range(len(provinces)):
        for province_2 in range(len(provinces)):
            if provinces[province_1]!=provinces[province_2]:
                
                if  (time_df['Province'] == provinces[province_1]).any() and (time_df['Province'] == provinces[province_2]).any():
                    
                    matrix[province_1][province_2] = str(get_diff(time_df[time_df['Province'] == provinces[province_1]].values[0][2],
                                                                  time_df[time_df['Province'] == provinces[province_2]].values[0][2]))
                else:
                    matrix[province_1][province_2] = "no data"                                                 
            else:
                matrix[province_1][province_2] = "/"
    over_years_matrix[year] = matrix
                   
    
```

```python
def print2d(array):
    for i in range(len(array[0])):
        for j in range(len(array[0])):
            print(array[i][j], end=' ')
        print()    
```
